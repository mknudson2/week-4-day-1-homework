```python
"""
Make a program that filters a list of strings and returns a list with only your friends name in it.

If a name has exactly 4 letters in it, you can be sure that it has to be a friend of yours! Otherwise, you can be sure he's not...

Ex: Input = ["Ryan", "Kieran", "Jason", "Yous"], Output = ["Ryan", "Yous"]

i.e.

friend ["Ryan", "Kieran", "Mark"] `shouldBe` ["Ryan", "Mark"]
Note: keep the original order of the names in the output.
"""

def friend(x):
    friends = [] #O(1) stating a list, no further actions needed
    for name in x: #O(N) iterates through each name in x
        if len(name) == 4: #O(1)  equating two objects in a given moment
            friends.append(name) #O(1) since we are adding to the end we do not need any extra steps
    return friends #O(1) 

# Total: #O(N)
```

```python
"""
If we list all the natural numbers below 10 that are multiples of 3 or 5, we get 3, 5, 6 and 9. The sum of these multiples is 23.

Finish the solution so that it returns the sum of all the multiples of 3 or 5 below the number passed in. Additionally, if the number is negative, return 0 (for languages that do have them).

Note: If the number is a multiple of both 3 and 5, only count it once.
"""


def solution(number):
    multiples = set()#O(1) --> use of set over list since a set can only contain unique values.
    
    if number <= 0: #O(1) comparing two numbers only and always
        return 0 #O(1)
    
    for value in range(number): #O(N) iterates through each number up to the given number
        if value%3==0 or value%5==0: #O(1) checks condition of a single item at a time
            multiples.add(value) #O(1) since sets are unsorted lists there is no concern for index shifting once added
    return sum(multiples) #O(1)

# Total: #O(N)
```

```python
"""
Return the number (count) of vowels in the given string.

We will consider a, e, i, o, u as vowels for this Kata (but not y).

The input string will only consist of lower case letters and/or spaces.
"""


def get_count(sentence):
    vowels = 'aeiou' #O(1)
    count = 0 #O(1)
    
    for char in sentence:#O(N) iterates through each character in a given string
        if char in vowels:#O(1)
            count += 1#O(1)
        
    return count#O(1)

# Total: #O(N)
```

```python
# 4th of July Problem

import random

def create_show(fireworks, show_time):
    fireworks.sort()

    show = []

    remaining_time = show_time


    while remaining_time > 0 and fireworks:

           # Select a random firework

           firework = random.choice(fireworks)


           if firework <= remaining_time:

               # Add the firework to the show

               show.append(firework)

               remaining_time -= firework

          else:

              # This firework is too long, remove it from consideration

              fireworks.remove(firework)

  return show
```
